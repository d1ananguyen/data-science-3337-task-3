```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv('../../data/cleaned_solarflare_dataset.csv')
df = df[df['total_counts'] >= 0]

# Convert date and start_time to datetime
df['datetime'] = pd.to_datetime(df['date'] + ' ' + df['start_time'], errors='coerce')
df = df.dropna(subset=['datetime', 'total_counts'])
df.set_index('datetime', inplace=True)

# Plot the flare intensity 
plt.figure(figsize=(14, 6))
plt.plot(df.index, df['total_counts'], label='Flare Intensity')
plt.title('Solar Flare Intensity Over Time')
plt.xlabel('Time')
plt.ylabel('Total Counts (612 keV)')
plt.grid(True)
plt.tight_layout()
plt.savefig('visuals/flare_intensity_over_time.png', dpi=300)
plt.show()


```

```python
global_mean = df['total_counts'].mean()
global_std = df['total_counts'].std()
df['z_total_counts'] = (df['total_counts'] - global_mean) / global_std


# 6H Charts
rolling_mean = df['total_counts'].rolling('6H').mean()
rolling_std = df['total_counts'].rolling('6H').std()

# Local Z-score and sudden changes
df['local_z'] = (df['total_counts'] - rolling_mean) / rolling_std
upper = df['local_z'].quantile(0.999)
lower = df['local_z'].quantile(0.001)
sudden_changes = df[(df['local_z'] > upper) | (df['local_z'] < lower)]
print("Num 6H", len(sudden_changes))


# Plot
plt.figure(figsize=(16, 6))

plt.plot(df.index, df['total_counts'], label='Flare Intensity', color='red', alpha=0.5)
plt.plot(rolling_mean.index, rolling_mean, label='Rolling Mean (6H)', color='blue')
plt.plot(rolling_std.index, rolling_std, label='Rolling Std Dev (6H)', color='green', alpha=0.5)

plt.scatter(sudden_changes.index, sudden_changes['total_counts'],
            color='crimson', edgecolor='black', s=50, label='Sudden Change', zorder=5)

plt.title('Solar Flare Intensity with Rolling Mean, Std Dev, and Sudden Change Annotations (6 Hour Rolling Window)')
plt.xlabel('Time')
plt.ylabel('Total Counts (Billions, 612 keV)')
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.savefig('visuals/flare_intensity_sudden_changes_6H.png', dpi=300)
plt.show()

plt.hist(df['local_z'].dropna(), bins=100, color='gray', alpha=0.6)
plt.axvline(upper, color='red', linestyle='--', label='99.9th Percentile')
plt.axvline(lower, color='blue', linestyle='--', label='0.01th Percentile')
plt.legend()
plt.title('Distribution of Local Z-Scores')
plt.xlabel('Z-Score')
plt.ylabel('Frequency')
plt.savefig('visuals/z_score_distribution_6H.png', dpi=300)
plt.show()


# 1D Charts
rolling_mean = df['total_counts'].rolling('1D').mean()
print(rolling_mean)
rolling_std = df['total_counts'].rolling('1D').std()
print(rolling_std)

# Local Z-score and sudden changes
df['local_z'] = (df['total_counts'] - rolling_mean) / rolling_std
upper = df['local_z'].quantile(0.999)
lower = df['local_z'].quantile(0.001)
sudden_changes = df[(df['local_z'] > upper) | (df['local_z'] < lower)]
print("Num 1D", len(sudden_changes))

# Plot
plt.figure(figsize=(16, 6))

plt.plot(df.index, df['total_counts'], label='Flare Intensity', color='red', alpha=0.5)
plt.plot(rolling_mean.index, rolling_mean, label='Rolling Mean (1D)', color='blue')
plt.plot(rolling_std.index, rolling_std, label='Rolling Std Dev (1D)', color='green', alpha=0.5)

plt.scatter(sudden_changes.index, sudden_changes['total_counts'],
            color='crimson', edgecolor='black', s=50, label='Sudden Change', zorder=5)

plt.title('Solar Flare Intensity with Rolling Mean, Std Dev, and Sudden Change Annotations (1 Day Rolling Window)')
plt.xlabel('Time')
plt.ylabel('Total Counts (Billions, 612 keV)')
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.savefig('visuals/flare_intensity_sudden_changes_1D.png', dpi=300)
plt.show()

plt.hist(df['local_z'].dropna(), bins=100, color='gray', alpha=0.6)
plt.axvline(upper, color='red', linestyle='--', label='99.9th Percentile')
plt.axvline(lower, color='blue', linestyle='--', label='0.01th Percentile')
plt.legend()
plt.title('Distribution of Local Z-Scores')
plt.xlabel('Z-Score')
plt.ylabel('Frequency')
plt.savefig('visuals/z_score_distribution_6H.png', dpi=300)
plt.show()


# 1W Charts
rolling_mean = df['total_counts'].rolling('7D').mean()
rolling_std = df['total_counts'].rolling('7D').std()

# Local Z-score and sudden changes
df['local_z'] = (df['total_counts'] - rolling_mean) / rolling_std
upper = df['local_z'].quantile(0.999)
lower = df['local_z'].quantile(0.001)
sudden_changes = df[(df['local_z'] > upper) | (df['local_z'] < lower)]
print("Num 1W", len(sudden_changes))


# Plot
plt.figure(figsize=(16, 6))

plt.plot(df.index, df['total_counts'], label='Flare Intensity', color='red', alpha=0.5)
plt.plot(rolling_mean.index, rolling_mean, label='Rolling Mean (1W)', color='blue')
plt.plot(rolling_std.index, rolling_std, label='Rolling Std Dev (1W)', color='green', alpha=0.5)

plt.scatter(sudden_changes.index, sudden_changes['total_counts'],
            color='crimson', edgecolor='black', s=50, label='Sudden Change', zorder=5)

plt.title('Solar Flare Intensity with Rolling Mean, Std Dev, and Sudden Change Annotations (1 Week Rolling Window)')
plt.xlabel('Time')
plt.ylabel('Total Counts (Billions, 612 keV)')
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.savefig('visuals/flare_intensity_sudden_changes_1W.png', dpi=300)
plt.show()

plt.hist(df['local_z'].dropna(), bins=100, color='gray', alpha=0.6)
plt.axvline(upper, color='red', linestyle='--', label='99.9th Percentile')
plt.axvline(lower, color='blue', linestyle='--', label='0.01th Percentile')
plt.legend()
plt.title('Distribution of Local Z-Scores')
plt.xlabel('Z-Score')
plt.ylabel('Frequency')
plt.savefig('visuals/z_score_distribution_1W.png', dpi=300)
plt.show()


# 1M Charts
rolling_mean = df['total_counts'].rolling('30D').mean()
rolling_std = df['total_counts'].rolling('30D').std()

# Local Z-score and sudden changes
df['local_z'] = (df['total_counts'] - rolling_mean) / rolling_std
upper = df['local_z'].quantile(0.999)
lower = df['local_z'].quantile(0.001)
sudden_changes = df[(df['local_z'] > upper) | (df['local_z'] < lower)]
print("Num 1M", len(sudden_changes))

# Plot
plt.figure(figsize=(16, 6))

plt.plot(df.index, df['total_counts'], label='Flare Intensity', color='red', alpha=0.5)
plt.plot(rolling_mean.index, rolling_mean, label='Rolling Mean (30D)', color='blue')
plt.plot(rolling_std.index, rolling_std, label='Rolling Std Dev (30D)', color='green', alpha=0.5)

plt.scatter(sudden_changes.index, sudden_changes['total_counts'],
            color='crimson', edgecolor='black', s=50, label='Sudden Change', zorder=5)

plt.title('Solar Flare Intensity with Rolling Mean, Std Dev, and Sudden Change Annotations (30 Day Rolling Window)')
plt.xlabel('Time')
plt.ylabel('Total Counts (Billions, 612 keV)')
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.savefig('visuals/flare_intensity_sudden_changes_1M.png', dpi=300)
plt.show()

plt.hist(df['local_z'].dropna(), bins=100, color='gray', alpha=0.6)
plt.axvline(upper, color='red', linestyle='--', label='99.9th Percentile')
plt.axvline(lower, color='blue', linestyle='--', label='0.01th Percentile')
plt.legend()
plt.title('Distribution of Local Z-Scores')
plt.xlabel('Z-Score')
plt.ylabel('Frequency')
plt.savefig('visuals/z_score_distribution_1M.png', dpi=300)
plt.show()


# 6M Charts
rolling_mean = df['total_counts'].rolling('180D').mean()
rolling_std = df['total_counts'].rolling('180D').std()

# Local Z-score and sudden changes
df['local_z'] = (df['total_counts'] - rolling_mean) / rolling_std
upper = df['local_z'].quantile(0.999)
lower = df['local_z'].quantile(0.001)
sudden_changes = df[(df['local_z'] > upper) | (df['local_z'] < lower)]
print("Num 6M", len(sudden_changes))

# Plot
plt.figure(figsize=(16, 6))

plt.plot(df.index, df['total_counts'], label='Flare Intensity', color='red', alpha=0.5)
plt.plot(rolling_mean.index, rolling_mean, label='Rolling Mean (180D)', color='blue')
plt.plot(rolling_std.index, rolling_std, label='Rolling Std Dev (180D)', color='green', alpha=0.5)

plt.scatter(sudden_changes.index, sudden_changes['total_counts'],
            color='crimson', edgecolor='black', s=50, label='Sudden Change', zorder=5)

plt.title('Solar Flare Intensity with Rolling Mean, Std Dev, and Sudden Change Annotations (180 Day Rolling Window)')
plt.xlabel('Time')
plt.ylabel('Total Counts (Billions, 612 keV)')
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.savefig('visuals/flare_intensity_sudden_changes_6M.png', dpi=300)
plt.show()

plt.hist(df['local_z'].dropna(), bins=100, color='gray', alpha=0.6)
plt.axvline(upper, color='red', linestyle='--', label='99.9th Percentile')
plt.axvline(lower, color='blue', linestyle='--', label='0.01th Percentile')
plt.legend()
plt.title('Distribution of Local Z-Scores')
plt.xlabel('Z-Score')
plt.ylabel('Frequency')
plt.savefig('visuals/z_score_distribution_6M.png', dpi=300)
plt.show()
```
